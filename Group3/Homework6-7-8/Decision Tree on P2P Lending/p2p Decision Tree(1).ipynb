```python
import numpy as np
import pandas as pd
from sklearn import tree
import graphviz
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score as acc_rate 
```

```python
p2p = pd.read_csv('p2p.csv', usecols = range(1,26)) # p2p = pd.read_csv('p2p.csv', index_col=0) 
p2p.head(10)
```

```python
x = p2p.drop('status', axis=1)
y = p2p.status

# randomly choose 1/3 of samples as testing data
np.random.seed(12345)
test_id = np.random.randint(0, p2p.shape[0], p2p.shape[0] // 3)

# split samples into training set and test set
train_x, train_y = x.drop(test_id), y.drop(test_id)
test_x, test_y= x.iloc[test_id, :], y[test_id]
```

```python
# initialize a decision tree classifier object with given arguments 
clf = tree.DecisionTreeClassifier(criterion='entropy', splitter='best', ) 
clf.fit(X=train_x.values, y=train_y.values)
```

```python
# make prediction
pred_y = clf.predict(test_x)
pred = pd.DataFrame({'y': test_y, 'pred_y': pred_y})
pred.head(20)
```

```python
print('\nAccuracy rate is:\n', acc_rate(test_y, clf.predict(test_x)))
```

```python
# visualizing the tree
dot_data = tree.export_graphviz(clf, out_file=None,
                                feature_names=x.columns.values, 
                                class_names=['0', '1'], 
                                filled=True, 
                                rounded=True, 
                                impurity=False, 
                                special_characters=True) 
graph = graphviz.Source(dot_data)
graph.render("p2p")
```

```python
max_depth = []
acc_gini = []
acc_entropy = []

for i in range(1,30):
    dtree = tree.DecisionTreeClassifier(criterion='gini', max_depth=i)
    dtree.fit(train_x, train_y)
    pred = dtree.predict(test_x)
    acc_gini.append(acc_rate(test_y, pred))
    dtree = tree.DecisionTreeClassifier(criterion='entropy', max_depth=i)
    dtree.fit(train_x, train_y)
    pred = dtree.predict(test_x)
    acc_entropy.append(acc_rate(test_y, pred))
    max_depth.append(i)
    
d = pd.DataFrame({'acc_gini':pd.Series(acc_gini), 'acc_entropy':pd.Series(acc_entropy), 'max_depth':pd.Series(max_depth)})

plt.plot('max_depth','acc_gini', data=d, label='gini')
plt.plot('max_depth','acc_entropy', data=d, label='entropy')
plt.xlabel('max_depth')
plt.ylabel('accuracy')
plt.legend()
```

```python
dtree = tree.DecisionTreeClassifier(criterion='entropy', max_depth=3)
dtree.fit(train_x, train_y)

pred = dtree.predict(test_x)
acc_rate(test_y, pred)
```

```python
# visualizing the tree
dot_data = tree.export_graphviz(dtree, out_file=None,
                                feature_names=x.columns.values, 
                                class_names=['0', '1'], 
                                filled=True, 
                                rounded=True, 
                                impurity=False, 
                                special_characters=True) 
graph1 = graphviz.Source(dot_data)
graph1.render("p2p optimized")
```
