```python
import os
import numpy as np
import pandas as pd
from datetime import datetime
path = "Multivariate//Data//Market Data//"
names = os.listdir(path)
names
```

```python
target_df = pd.read_csv("DCOILBRENTEU.csv",index_col = 0)
target_df.index = pd.to_datetime(target_df.index)
target_df.index
```

```python
def process_df(df1,name):
    df = df1.copy()
    df = df.iloc[::-1]
    df = df.dropna()
    df.index.name = None
    if len(df.columns) > 1:
        df.columns = [name,name+'vol']
    else:
        df.columns = [name]
    df.index = pd.to_datetime(df.index)
    dates = pd.date_range(df.first_valid_index().date(),df.last_valid_index().date(),freq = 'D')
    df1 = pd.DataFrame(index=dates)
    df = pd.concat([df1,df],axis = 1)
    df = df.sort_index()
    df = df.ffill()
    return df
```

```python
df1 = pd.read_excel(path+names[0],index_col = 0,header = 6)

df1 = process_df(df1,names[0][:-5])
df = pd.read_excel(path+names[1],index_col = 0,header = 6)

df = process_df(df,names[1][:-5])
df1.index[-1]
```

```python
target_df = target_df.ffill()
target_df
```

```python
target_df = process_df(target_df,'WFI')
df_all = target_df#pd.DataFrame()
for i in [-6,-5,-3,-1,7]:

    df = pd.read_excel(path+names[i],index_col = 0,header = 6)

    df = process_df(df,names[i][:-5])
    print(df.index)
    df_all = pd.concat([df_all,df],axis = 1)

df_all
```
